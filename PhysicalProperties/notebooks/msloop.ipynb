# Exploring MSLOOP data

```python
file = './data/input/MSLOOP/400-U1603A-1H-1_20230824145717.MS'
df = utils.read_instrument_file(file, as_dataframe=True)
df
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))

x = df["magnetic_susceptibility"].astype(float)
y = df['offset'].astype(float)

ax.scatter(
    x=x,
    y=y,
    label='data',
    color='blue',
    edgecolor='black'
)

windowsize = 3
ax.plot(
    utils.convolve(x,windowsize),
    y,
    label=f'avg {windowsize} pts'
)


ax.invert_yaxis()
ax.set_ylabel("offset (cm)")
ax.set_xlabel("magnetic susceptibility")
ax.set_title(f"{df['labelid'][0]}")
ax.legend()
plt.plot()
```
